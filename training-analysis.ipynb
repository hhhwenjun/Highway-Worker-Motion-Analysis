```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("worker5_feature.csv", sep=',')
```

```python
df.drop(columns = ["Unnamed: 0", "Unnamed: 0.1"],inplace=True)
```

```python
df
```

```python
df.info()
```

```python
X=df.drop(['label'],axis=1)
y=df['label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

Feature Importance

```python
from sklearn.feature_selection import SelectPercentile
import numpy as np
select = SelectPercentile(percentile=70)
select.fit(X_train, y_train)
# transform training set:
X_train_selected = select.transform(X_train)
print('Feature scores：',select.scores_)
print('Dimensions before filtering：',X_train.shape)
print('Dimensions after filtering：',X_train_selected.shape)
np.argsort(select.scores_) #select features based on score
```

LR

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report

clf = GridSearchCV(
    estimator = LogisticRegression (),
    param_grid = {'penalty': ['l1','l2'],
                  'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]},
    cv=StratifiedKFold(n_splits=3).split(X_train, y_train)
)

clf.fit(X_train,y_train)

print(clf.best_params_)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

SVM

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC 
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report

# defining parameter range 
clf = GridSearchCV(
    estimator = SVC(),
    param_grid = {'C': [0.1, 1, 10, 100, 1000],  
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
              'kernel': ['rbf']},
    refit = True, 
    verbose = 3,
    cv=StratifiedKFold(n_splits=3).split(X_train, y_train)
)

clf.fit(X_train,y_train)

print(clf.best_params_)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

Naive Bayes

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB 
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report

# defining parameter range 
clf = GridSearchCV(
    estimator = GaussianNB(),
    param_grid = {},
    refit = True, 
    verbose = 3,
    cv=StratifiedKFold(n_splits=3).split(X_train, y_train)
)

clf.fit(X_train,y_train)

print(clf.best_params_)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

K-NN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report

k_range = list(range(1,30))
weight_options = ["uniform", "distance"]

clf = GridSearchCV(
    estimator = KNeighborsClassifier(),
    param_grid = dict(n_neighbors = k_range, weights = weight_options),
    cv=StratifiedKFold(n_splits=3).split(X_train, y_train)
)

clf.fit(X_train,y_train)

print(clf.best_params_)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report

clf = GridSearchCV(
    estimator = DecisionTreeClassifier(),
    param_grid ={'max_depth': np.arange(3, 10)},
    cv=StratifiedKFold(n_splits=3).split(X_train, y_train)
)

clf.fit(X_train,y_train)

print(clf.best_params_)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

RF

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report

clf = GridSearchCV(
    estimator = RandomForestClassifier(),
    param_grid = {'n_estimators': [10, 20, 50],
                  'max_depth': [4,5,6,7]},
    cv=StratifiedKFold(n_splits=3).split(X_train, y_train)
)


clf.fit(X_train,y_train)

print(clf.best_params_)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import lightgbm as lgb
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report

clf = GridSearchCV(
    estimator = lgb.LGBMClassifier(class_weight='balanced'),
    param_grid = {'boosting_type':['gbdt','dart','rf'],'n_estimators': [10, 20, 50,80],
                  'max_depth': [3,4,5,6,10,15],
                  'subsample':[0.6, 0.7, 0.8]},
    cv=StratifiedKFold(n_splits=3).split(X_train, y_train)
)

clf.fit(X_train,y_train)

print(clf.best_params_)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import joblib
joblib.dump(clf.best_estimator_, 'LBGM_ScanSSD.pkl')
```
